```python
import pandas as pd
```

```python
data_link = "https://github.com/fedhere/DSPS_FBianco/raw/refs/heads/main/data/earthquakes.csv"
# link found https://github.com/fedhere/DSPS_FBianco/blob/main/data/earthquakes.csv
earthquake = pd.read_csv(data_link, sep= r"\s+")
earthquake = earthquake[["#YYY/MM/DD", "HH:mm:SS.ss", "MAG"]]
earthquake.dtypes
```

```python
earthquake = earthquake.rename(columns = {"#YYY/MM/DD" : "date",
                             "HH:mm:SS.ss": "time",
                             "MAG": "mag"})
# remember pd doesn't change values need to repoint to earthquake or do inplace
earthquake.head()
```

```python
earthquake.describe()
```

```python
pd.to_datetime("01:30:57.52")
```

```python
earthquake["new_date"] = earthquake["date"].map(pd.to_datetime) # takes a second because 30,000 values
```
